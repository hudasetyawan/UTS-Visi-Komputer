```python
# Install library jika diperlukan
!pip install opencv-python-headless matplotlib

# Import library
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Upload gambar daun
from google.colab import files
uploaded = files.upload()

# Load gambar
for filename in uploaded.keys():
    img = cv2.imread(filename)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Tampilkan gambar RGB dan Grayscale
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title('Gambar RGB')
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('Gambar Grayscale')
plt.imshow(img_gray, cmap='gray')
plt.axis('off')
plt.show()

# Proses CLAHE (perbaikan kontras)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_clahe = clahe.apply(img_gray)

# Thresholding Global (Otsu)
_, thresh_img = cv2.threshold(img_clahe, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Penajaman Tepi dengan Laplacian of Gaussian (LoG)
# Pertama, Gaussian Blur
blurred = cv2.GaussianBlur(thresh_img, (5,5), 0)
# Kemudian, Laplacian
log_edges = cv2.Laplacian(blurred, cv2.CV_64F)
log_edges = np.uint8(np.absolute(log_edges))

# Tampilkan semua hasil
plt.figure(figsize=(18,6))
plt.subplot(1,3,1)
plt.title('CLAHE + Thresholding')
plt.imshow(thresh_img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title('Hasil Laplacian of Gaussian (LoG)')
plt.imshow(log_edges, cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.title('Histogram Intensitas')
plt.hist(img_gray.ravel(), 256, [0,256])
plt.xlabel('Intensitas')
plt.ylabel('Jumlah Pixel')
plt.show()
```
